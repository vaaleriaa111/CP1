```python
import pandas as pd 
```

```python
dataF="https://www.cia.gov/the-world-factbook/field/total-fertility-rate/country-comparison" 
```

```python
fertility=pd.read_html(dataF) #escarpear
fertility #refiere a un link de la página (muchas tablas)
```

```python
type(fertility) #para saber que tipo de data es fertility 
```

```python
fertility[0] #Utilizando el [0] se escoge la primera lista
```

```python
#1. Cambiar nombres de columnas 
fertilityOK=fertility[0].rename(columns={"children born/woman":"hijosxmujer"})
```

```python
fertilityOK
```

```python
#2. Asegurarme que no haya espacios entre nombres de columnas 
fertilityOK.columns=fertilityOK.columns.str.replace(' ','')
```

```python
fertilityOK
```

```python
fertilityOK=fertilityOK.rename(columns={"DateofInformation": "fecha_info"})
```

```python
fertilityOK
```

```python
#3. Mantener solo el año en la columna fecha_info 
```

```python
fertilityOK.fecha_info=fertilityOK.fecha_info.str.split(' ',expand=True)[0]
```

```python
fertilityOK
```

```python
#4. Detectar Leading or Trailing spaces 
fertilityOK.Country[fertilityOK.Country.str.contains(pat=r'[^\w\s]',regex=True)]
```
